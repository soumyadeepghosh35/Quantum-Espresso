```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Apr 16 15:07:43 2019

@author: soumyadeep
"""


##################################################
# Band Structure DFT+TB
##################################################


import numpy as np
import matplotlib.pyplot as plt
import csv
import shutil
import glob, os, os.path
from matplotlib import rc,rcParams
import pandas as pd




#information about system
'ELEMENT= CsCa2As4Fe4F2'
nkpts=301           #no of kpoints 
nbnd=77           # number of bands
nwan=20             # number of wannier functions
Efermi1=9.5239     # fermi energy in eV  DFT
Efermi2=9.5239     # fermi energy in eV  WANNIER







shutil.rmtree('/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/bands_info_DFT')
shutil.rmtree('/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/bands_info_TB')
os.remove('/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/DFT_BAND.csv')
os.remove('/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/TB_BAND.csv')
#filelist = glob.glob(os.path.join('/home/soumyadeep/Python/QE/Fe/', "*.csv"))
#for f in filelist:
#    os.remove(f)
#os.remove("/home/soumyadeep/Python/QE/Fe/*.csv")
os.mkdir('/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/bands_info_DFT')
os.mkdir('/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/bands_info_TB')





# this part extract data from DFT band output file
z = np.loadtxt('/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/CsCa2F2.band.dat.gnu')    # original DFT bandstructure file
for i in range(len(z)):
    a=z[i][0]               # a=k_point
    b=z[i][1]-Efermi1       # b=energy
    #print(a,b)
    with open("/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/DFT_BAND.csv","a") as output_file:
      output_fileWriter=csv.writer(output_file)
      output_fileWriter.writerow([a,b])
    output_file.close()  
    

# this part extract data from WANNIER band output file
z = np.loadtxt('/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/CsCa2F2_band.dat')       # original TB bandstructure file
for i in range(len(z)):
    A=z[i][0]               # a=k_point
    B=z[i][1]-Efermi2       # b=energy
    #print(A,B)
    with open("/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/TB_BAND.csv","a") as output_file:
      output_fileWriter=csv.writer(output_file)
      output_fileWriter.writerow([A,B])
    output_file.close()    




# this part will separate data for each DFT band
csvfile = open('/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/DFT_BAND.csv', 'r').readlines()
filename = 1
for i in range(len(csvfile)):
     if i % nkpts == 0:
         open('/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/bands_info_DFT/' + str(filename) + 'DFT_band.csv', 'w+').writelines(csvfile[i:i+nkpts])
         filename += 1  



# this part will separate data for each WANNIER/TB band
csvfile = open('/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/TB_BAND.csv', 'r').readlines()
filename = 1
for i in range(len(csvfile)):
     if i % nkpts == 0:
         open('/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/bands_info_TB/' + str(filename) + 'TB_band.csv', 'w+').writelines(csvfile[i:i+nkpts])
         filename += 1 
         


# this part will compare both the band structures
fig,ax = plt.subplots()
path = "/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/bands_info_DFT/*DFT_band.csv"   #DFT input file
for fname in glob.glob(path):    
    k, E= np.loadtxt(fname, delimiter=',', unpack=True)
    k=k/max(k)
    pDFT,=plt.plot(k, E, c='navy', lw=1.5, label='DFT')



path = "/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/bands_info_TB/*TB_band.csv"   # TB input file
for fname in glob.glob(path):    
    k, E= np.loadtxt(fname, delimiter=',', unpack=True)
    k=k/max(k)
    pTB,=plt.plot(k, E, c='r',  lw=0.8 , ls='--', label='TB')



    
plt.ylim(-1.0,1.0)
plt.xlim(min(k),max(k))  
#plt.yticks(np.arange(-1, 1.2, 0.2))         
plt.axhline(y=0.0, color='g',ls='--',lw=1.0)
plt.axvline(x=0.333,color='k', ls='--',lw=1.0)
plt.axvline(x=0.666,color='k', ls='--',lw=1.0)
plt.xticks([0.0,0.333,0.666,1.0], ['$\Gamma$', 'X', '$\Gamma$','X'],fontsize=18)
plt.annotate(r'$\rm E_{F}$', xy=(0,0), xytext=(1.01,-0.05),fontsize=18,weight="bold")
plt.ylabel(r'$\rm E-E_{F}(eV)$',fontsize=18)
plt.xticks(fontweight='bold')
plt.rc('font', weight='bold')
plt.rc('ytick.major')
plt.rc('ytick')
leg=plt.legend((pDFT,pTB), ('DFT','TB'), loc="upper right", bbox_to_anchor=(0.46,1.02), frameon=False, framealpha=1.0, facecolor= 'white')
plt.suptitle(r'$\rm CsCa_2As_4Fe_4F_2$ Band Structure Comparision',y=0.92,weight='bold',size=10)
fig.set_size_inches(5,4)
plt.savefig("/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/fig1.pdf",bbox_inches='tight', dpi=2500)
plt.savefig("/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/CsCa244F2-DFT+TB_BS.pdf",bbox_inches='tight', dpi=4500)
plt.savefig("/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/CsCa244F2-DFT+TB_BS.eps",bbox_inches='tight', dpi=4500)
plt.savefig("/media/soumyadeep/Soumya2/Python/12442/CsCa244F2/CsCa244F2-DFT+TB_BS.jpeg",bbox_inches='tight', dpi=1500)  
plt.show()  

print('sucessful') 

     
```
